#### DATA VISUALISATION 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))
plt.title("DASHBOARD")
plt.subplot(2,3,1)
sns.countplot(x="Survived",color="green",data=df)
plt.subplot(2,3,2)
sns.countplot(x="Sex",hue="Survived",data=df)
a= df["Survived"].value_counts()
plt.subplot(2,3,3)
plt.pie(a.values, labels=a.index, autopct="%1d%%")
b=df["Pclass"].value_counts()
plt.subplot(2,3,4)
plt.bar(b.index, b.values,color="red")
plt.subplot(2,3,5)
sns.scatterplot(x="Age",y="Fare",hue="Survived",data=df)
plt.subplot(2,3,6)
sns.boxplot(x="Fare",color="purple",data=df)
```

## DATA STORY TELLING

#### 1. Survival Count plot:- this graph shows us how many passengers survived and how many did not survive. 
        we can see that more passengers did not survive (dead)
#### 2. Gender VS Survival plot:- this graph shows us survival based on gender 
        we can see that female passengers had a higher survival rate
#### 3. Survival pie chart:- this shows the percentage of who survived and who did not survive
        it gives quick overall review 
        we can see that first class passengers survived more in the accident 
#### 4. Passenger class barchar:- this shows number of passengers in each class(1st,2nd and 3rd)
        we can see most passengers were travelling in 3rd class
#### 5. Age VS Fare scatter plot:- this shows relationship between passenger and ticket fare
        here colors indicates survived or not 
        we can see the passengers who paid more are the people who survived more 
#### 6. Fare box plot:- this shows the distribution of ticket fares 
        we can see most of the passengers paid lower fairs. 
        the passengers who paid more are seen as outliers in the plot.
